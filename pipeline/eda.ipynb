## Packages and Assets

```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_preprocessed = pd.read_csv('../assets/data/splits/train/preprocessed.csv')
val_preprocessed = pd.read_csv('../assets/data/splits/val/preprocessed.csv')
test_preprocessed = pd.read_csv('../assets/data/splits/test/preprocessed.csv')
df_preprocessed = pd.concat([train_preprocessed, val_preprocessed, test_preprocessed])
df_preprocessed
```

```python
petr4 = df_preprocessed.loc[df_preprocessed['label'] == 1]
vale3 = df_preprocessed.loc[df_preprocessed['label'] == 2]
itub4 = df_preprocessed.loc[df_preprocessed['label'] == 3]
outras = df_preprocessed.loc[df_preprocessed['label'] == 0]
```

## Functions

```python
stopwords = ["da", "meu", "em", "voce", "de", "ao", "os", "<NUM>", "<num>", "num", "NUM", 'acoe', 'acoes', 'milhoe',' bilhoe', 'milhoes', 'cento', 'diz', 'bilhoes']
```

```python
summary_petr4 = " ".join(str(s) for s in petr4['title'].values)
summary_vale3 = " ".join(str(s) for s in vale3['title'].values)
summary_itub4 = " ".join(str(s) for s in itub4['title'].values)
summary_outras = " ".join(str(s) for s in outras['title'].values)
```

## PETR4

### Wordcloud

```python
wordcloud_petr4 = WordCloud(collocations=False, stopwords=stopwords, background_color='white', width=1600, height=800).generate(summary_petr4)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud_petr4)
plt.axis('off')
plt.title("Palavras que mais aparecem em notícias da classe Petrobras (Classe 1)")
plt.show()
```

### See news that doesnt have the word "petrobras" or "petr4"

```python
petr4
```

```python
petr4.loc[(petr4['title'].str.contains('petrobras') == False) & (petr4['title'].str.contains('petr4') == False)]
```
