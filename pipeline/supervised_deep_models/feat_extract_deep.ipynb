## Word2vec

```python
import pandas as pd
from sklearn.manifold import TSNE
import plotly.express as px
import numpy as np
from gensim.models import Word2Vec
```

### Functions

### Creating or Loading Word2Vec Model

```python
hasModel = False
tokenized_corpus = []
if not hasModel:

    for title in train_pp['title']:
        tokenized_corpus.append(title.split())
    model = Word2Vec(sentences=tokenized_corpus, vector_size=300, window=20, min_count=5,workers =4)
else:
    print("Model Exists!")
    model = Word2Vec.load("../../assets/deep_assets/word2vec.model")
```

```python
model.wv.most_similar('petr4',topn=10)
```

```python
word_embeddings = []
vocab = []

for word in model.wv.index_to_key:
    vocab.append(word)
    word_embeddings.append(model.wv[word])
```

```python
emb_df = pd.DataFrame(word_embeddings)
emb_df = emb_df.assign(nome=vocab)
emb_df
```

### Visualizing Word Embeddings using TSNE and PCA

```python
tsne = TSNE(n_components=2,perplexity=40, n_iter=2500, random_state=23, learning_rate=100,metric='euclidean',init='pca')
X_tsne = tsne.fit_transform(np.array(word_embeddings))
X_tsne[1:4,:]
emb_df['x'] = X_tsne[:,0]
emb_df['y'] = X_tsne[:,1]
fig = px.scatter(emb_df,x='x',y='y',hover_data={'x':False, 'y':False,'nome':True})
fig.show()
```

## Feat Extraction

### Packages

```python
import pandas as pd
from keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
import numpy as np
```

### Functions

```python
def createVocabulary(corpus):
    """
    - Cria vocab (palavra e sua respectiva frequência no corpus)
    - Cria tokens (palavras do corpus)

    :param corpus: lista de string
    :return vocab, tokens, vocab_size:
    """
    tokens = []  # {'deeds', 'old', ...} 71666
    vocab = {}  # {'deeds': 2, 'old': 20', ...} 17971
    for text in corpus:
        for token in text.split():
            tokens.append(token)
            if token in vocab:
                vocab[token] += 1
            else:
                vocab[token] = 1
    vocab_size = len(vocab)
    return tokens, vocab, vocab_size
```

```python
def findMaxLen(sequence):
    max_len = 0
    for text in sequence:
        if len(text) > max_len:
            max_len = len(text)
    return max_len

def findAverageLen(sequence):
    total_len = 0
    for text in sequence:
        total_len += len(text)
    return total_len / len(sequence)
```

### Selecting Features

```python
X_train = train_pp.title
y_train = train_pp.label

X_val = val_pp.title
y_val = val_pp.label

X_test = test_pp.title
y_test = test_pp.label
```

### Tokens and Vocab

```python
tokens, vocab, vocab_size = createVocabulary(X_train)
len(vocab)
```

### Numericalization

```python
tokenizer = Tokenizer(num_words=vocab_size, oov_token='<OOV>')
tokenizer.fit_on_texts(X_train)
word_index = tokenizer.word_index

train_sequences = tokenizer.texts_to_sequences(X_train)
val_sequences = tokenizer.texts_to_sequences(X_val)
test_sequences = tokenizer.texts_to_sequences(X_test)
```

### Padding

```python
max_len = findMaxLen(train_sequences)
max_len_pad = int(max_len/2)
max_len_pad
```

```python
train_padded = pad_sequences(train_sequences, maxlen=max_len_pad, padding='post', truncating='post')

val_padded = pad_sequences(val_sequences, maxlen=max_len_pad, padding='post', truncating='post')

test_padded = pad_sequences(test_sequences, maxlen=max_len_pad, padding='post', truncating='post')
```

```python
train_padded = np.insert(train_padded, max_len_pad, y_train, 1)
val_padded = np.insert(val_padded, max_len_pad, y_val, 1)
test_padded = np.insert(test_padded, max_len_pad, y_test, 1)
```
